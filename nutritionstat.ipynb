```python
from scipy.stats import ttest_ind
```

```python
# Select protein values for each category
fruits = df[df['category'] == 'Fruits']['protein']
cereals = df[df['category'] == 'Cereals']['protein']

# Perform independent t-test
t_stat, p_val = ttest_ind(fruits, cereals)

# Output results
print('t-statistic:', t_stat)
print('p-value:', p_val)
```

```python
#anova
from scipy.stats import f_oneway

fruits = df[df['category'] == 'Fruits']['calories']
cereals = df[df['category'] == 'Cereals']['calories']
snacks = df[df['category'] == 'Snacks']['calories']

f_stat, p_val = f_oneway(fruits, cereals, snacks)
print('F-statistic:', f_stat, 'p-value:', p_val)

```

```python
from statsmodels.stats.weightstats import ztest

# Example: calories
z_stat, p_val = ztest(df['calories'], value=df['calories'].mean())
print('z-statistic:', z_stat, 'p-value:', p_val)
```

```python
from scipy.stats import chi2_contingency

# Create a high/low protein category
df['protein_level'] = ['High' if x > df['protein'].median() else 'Low' for x in df['protein']]

# Create a contingency table
cont_table = pd.crosstab(df['category'], df['protein_level'])
print(cont_table)
```

```python
chi2, p, dof, expected = chi2_contingency(cont_table)

print('Chi2 statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
```

p < 0.05  Significant association between category and protein level.

p  0.05  No significant association.
