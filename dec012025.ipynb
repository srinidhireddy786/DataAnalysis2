```python
mode_cols = ['Gender', 'Married', 'Dependents', 'Education',
             'Self_Employed', 'Credit_History', 'Property_Area']

df[mode_cols] = df[mode_cols].fillna(df[mode_cols].mode().iloc[0])
```

```python
median_cols = ['ApplicantIncome', 'LoanAmoun', 'LoanAmount', 'Loan_Amount_Term']

df[median_cols] = df[median_cols].fillna(df[median_cols].median())
```
